## step 2 make a tiny room

```python
room = ["D","C","D","D","C"]
def show_room(room):
    picture = ""
    for spot in room:
        if spot == "D":
            picture += "💩"
        else:
            picture += "✨"
    print(picture)

print("our room right now:")
show_room(room)
```

## Step 3 teach the robot one simple rule

```python
def clean_spot(spot):
    if spot == "D":
        print("cleaning the spot.....")
        return "C"
    else:
        print("This spot is already clean!")
        return "C"
result = clean_spot("D")
print("The robot is looked at a dirty spot 💩 and made it:",result)
print(result)
```

## step 4 send the robot aroung the room

```python
print("BEFORE - the dirty room:")
show_room(room)
print()

for i in range(len(room)):
    room[i] = clean_spot(room[i])
    print("After cleaning spot number", i+1, ":")
    show_room(room)

print()
print("AFTER - all done! ✨🤖✨")
show_room(room)
```
